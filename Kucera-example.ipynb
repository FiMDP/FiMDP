### Show function
The function show can pass parameters to rendering of the consMDP. 

* `m` -- displays `minInitCons` (orange color) if already computed
* `M` -- displays `minInitCons` (orange color) always (computes it if necessary)
* `s` -- displays `sufficientLevels` (red color) if already computed
* `S` -- displays `sufficientLevels` (red color) always (computes it if necessary)
* `p` -- displays `positiveReachability` (blue color) if already computed
